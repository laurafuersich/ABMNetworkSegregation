# plot segregation curve

```python
data_segcurve=pd.DataFrame(columns=['distance','segregation','sample_id','iteration'])


reporters=pd.read_csv('reporters.csv')
parameters=pd.read_csv('parameters_sample.csv')

df = pd.merge(reporters, parameters, on="sample_id") 


sample_ids = df['sample_id'].unique()
print(sorted(sample_ids))

iterations = df['iteration'].unique()
print(sorted(iterations))

for x in sample_ids:
    for i in iterations:


       # subset = df[df["iteration"] == i & df["sample_id"==x]]


        subset=df[(df['iteration']==i) & (df['sample_id']==x)]
        print(subset)


        distance_values=subset['distance_values'].iloc[0]
        seg_curve=subset['seg_curve'].iloc[0]


        seg_curve=seg_curve.replace('[', '')
        seg_curve=seg_curve.replace(']', '')

        distance_values=distance_values.replace('[', '')
        distance_values=distance_values.replace(']', '')



        list_dist = distance_values.split(",")


        list_seg = seg_curve.split(",")

        data = pd.DataFrame({'distance':list_dist,
                        'segregation':list_seg })

        data['sample_id'] = x
        data['iteration'] = i
        print(data)

        data_segcurve=pd.concat([data, data_segcurve])

        #return(data_segcurve)

print(data_segcurve)


data_segcurve = pd.merge(data_segcurve, parameters, on="sample_id") 

```

```python

data_segcurve['segregation'] = data_segcurve['segregation'].astype('float64')
#print(data_segcurve.dtypes)

# subset to cluster size ==1 
plot_data=data_segcurve[(data_segcurve['clustering_threshold']==1)]

sns.lineplot(
    data=plot_data,
    #hue='clustering_threshold',
    y='segregation',
    x='distance',
    errorbar=(None)
    
) 
# subset to cluster size ==0.8
plot_data=data_segcurve[(data_segcurve['clustering_threshold']==0.8)]

sns.lineplot(
    data=plot_data,
    #hue='clustering_threshold',
    y='segregation',
    x='distance',
    errorbar=(None)
    
) 
# subset to cluster size ==0.6
plot_data=data_segcurve[(data_segcurve['clustering_threshold']==0.6)]

sns.lineplot(
    data=plot_data,
    #hue='clustering_threshold',
    y='segregation',
    x='distance',
    errorbar=(None)
    
) 
# subset to cluster size ==0.4
plot_data=data_segcurve[(data_segcurve['clustering_threshold']==0.4)]

sns.lineplot(
    data=plot_data,
    #hue='clustering_threshold',
    y='segregation',
    x='distance',
    errorbar=(None)
    
) 

plt.xlabel("Distance")
plt.ylabel("H Index")

plt.show()
```
